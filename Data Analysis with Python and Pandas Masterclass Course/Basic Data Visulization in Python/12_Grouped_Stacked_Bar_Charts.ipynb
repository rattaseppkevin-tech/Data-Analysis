```python
import pandas as pd

sales = pd.read_csv(r"C:\Users\Kevin\Desktop\transactions.csv", parse_dates=["date"])
```

```python
store_40s = list(range(40, 50))

(
sales
.query("store_nbr in @store_40s and date.dt.month in [1, 2, 3]")
.pivot_table(index=sales["date"].dt.month,
             columns= sales["store_nbr"],
             values= "transactions",
             aggfunc= "sum").apply(lambda x: x * 100 / sum(x), axis=1)
.plot.bar(stacked=True).legend(bbox_to_anchor=(1, 1))
)
```
