```python
import pandas as pd

sales = pd.read_csv(r"C:\Users\Kevin\Desktop\transactions.csv", parse_dates=["date"])
```

```python
(sales.query("store_nbr in [44, 47] and date.dt.year ==2014")
 .pivot_table(index = "date", columns = "store_nbr")
 .droplevel(0, axis=1)
 .plot(color=["#45A8A8", "#b927da"],
       xlabel = "Date",
       ylabel = "Transactions",
       #cmap = "tab10",
       alpha =.6,
       style = ["-", "--"],
       grid = True
 )).legend(bbox_to_anchor=(1, 1))
```
