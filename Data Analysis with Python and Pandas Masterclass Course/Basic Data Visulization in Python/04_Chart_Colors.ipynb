```python
import pandas as pd

sales = pd.read_csv(r"C:\Users\Kevin\Desktop\transactions.csv", parse_dates=["date"])
```

```python
sales_44 = sales.loc[sales["store_nbr"] == 44, ["date", "transactions"]]
sales_44.set_index("date").plot(
    title = "Store 44 Transactions 2013-2017",
    xlabel = "Date",
    ylabel = "Daily Transactions",
    color = "red"
)
```
