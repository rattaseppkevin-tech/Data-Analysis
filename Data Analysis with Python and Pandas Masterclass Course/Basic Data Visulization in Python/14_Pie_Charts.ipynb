```python
import pandas as pd

sales = pd.read_csv(r"C:\Users\Kevin\Desktop\transactions.csv", parse_dates=["date"])
```

```python
sales.head()
```

```python
store_40s = list(range(40, 50))

(
sales
.query("store_nbr in @store_40s and date.dt.month in [1, 2, 3]")
.groupby("store_nbr")
.agg({"transactions": "sum"})
.plot.pie(y="transactions")
.legend(bbox_to_anchor=(1, 1))
)
```

```python
(sales
.query("store_nbr in @store_40s")
.pivot_table(index=sales["date"].dt.month,
             columns="store_nbr",
             values="transactions",
             aggfunc="sum")
.plot.scatter(x=44, y=47)
)
```
