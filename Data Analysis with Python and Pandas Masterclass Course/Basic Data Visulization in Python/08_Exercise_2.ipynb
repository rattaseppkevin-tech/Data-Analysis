```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.clf()
sns.set_style("darkgrid")

oil = pd.read_csv(r"C:\Users\Kevin\Desktop\oil.csv", parse_dates=["date"])
oil.columns = ["dates", "prices"]
oil["prices"] = pd.to_numeric(oil["prices"], errors="coerce")
oil["dates"] = pd.to_datetime(oil["dates"])
```

```python
oil.rename(columns={"prices": "Oil Prices"}).plot(
    x = "dates",
    y = "Oil Prices",
    color = "black",
    xlabel = "Date",
    ylabel = "USD Price",
    title = "Oil Price 2013-2017",
    figsize = (10, 5),
    grid=True
)
```
