```python
import pandas as pd

sales = pd.read_csv(r"C:\Users\Kevin\Desktop\transactions.csv", parse_dates=["date"])
```

```python
(sales
.query("store_nbr in [2, 3]")
.pivot_table(index=sales["date"],
             columns="store_nbr",
             values="transactions",
             aggfunc="sum")
.plot.scatter(x=2, y=3)
)
```

```python
sales["month"] = sales["date"].dt.month

store_pivot = (
    sales
    .query("store_nbr in [2, 3]")
    .pivot_table(index="date", 
                 columns="store_nbr", 
                 values=["transactions", "month"])
)

store_pivot.plot.scatter(    
    x=("transactions", 2), 
    y=("transactions", 3), 
    c=("month", 2), 
    colormap="viridis", 
    figsize=(10, 8),
    title="Pood 2 vs Pood 3"
)
```
