```python
import pandas as pd

sales = pd.read_csv(r"C:\Users\Kevin\Desktop\transactions.csv", parse_dates=["date"])
```

```python
sales.query("store_nbr in [1, 2, 3, 4]").pivot(
    index="date", columns="store_nbr", values="transactions").plot(
    title=["Pood 1", "Pood 2", "Pood 3", "Pood 4"],
    xlabel= "date",
    ylabel= "transactions",
    alpha =.6,
    grid = True,
    subplots=True,
    layout = (2, 2),
    sharey=True,
    figsize=(10,5)
)
```
